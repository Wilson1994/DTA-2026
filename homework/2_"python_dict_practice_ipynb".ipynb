# Розділ 1. Словники (`dict`)

**Швидке нагадування:**

Словник — це колекція пар `ключ : значення`, створюється у фігурних дужках:
```python
student = {"name": "Олена", "age": 20, "city": "Київ"}
```

**Основні операції:**
- Доступ: `student["name"]`
- Додавання/зміна: `student["email"] = "olena@example.com"`
- Видалення: `del student["age"]`
- Перевірка: `if "name" in student:`
- Безпечне отримання: `student.get("phone")` — повертає `None`, якщо ключа немає
- Довжина: `len(student)` — кількість пар

**Перебір у циклі:**
- `for key in d.keys():` — перебір ключів
- `for value in d.values():` — перебір значень
- `for key, value in d.items():` — перебір пар

Дано словник з товарами магазину, де значення — це інший словник з ціною і кількістю на складі. Виконайте:
1. Виведіть загальну вартість всього товару на складі (ціна × кількість для кожного).
2. Виведіть назви товарів, яких залишилось менше 20 штук.
3. Виведіть найдорожчий товар.

````python
inventory = {
    "Ноутбук":   {"price": 18500, "stock": 12},
    "Телефон":   {"price": 9200,  "stock": 35},
    "Планшет":   {"price": 7800,  "stock": 8},
    "Навушники": {"price": 850,   "stock": 120},
    "Клавіатура":{"price": 1200,  "stock": 45},
    "Мишка":     {"price": 450,   "stock": 18}
}
# Ваш код:

total_value = 0

for product, info in inventory.items():
    total_value += info["price"] * info["stock"]

print("Загальна вартість складу:", total_value, "грн")

print("\nТовари з залишком менше 20:")

for product, info in inventory.items():
    if info["stock"] < 20:
        print(product, "-", info["stock"], "шт")

max_name = ""
max_price = 0

for product, info in inventory.items():
    if info["price"] > max_price:
        max_price = info["price"]
        max_name = product

print("\nНайдорожчий товар:", max_name)
print("Ціна:", max_price, "грн")
````

### Завдання 22. Телефонна книга з пошуком ⭐⭐⭐

Створіть словник-телефонну книгу з 5 контактами. Напишіть програму, яка у циклі `while`:
1. Запитує у користувача дію: `пошук`, `додати`, `видалити`, `вихід`
2. Залежно від вибору — виконує відповідну дію
3. Завершується, коли користувач введе `вихід`

> 💡 **Підказка:** Використайте `while True:` (нескінченний цикл) і умову завершення з `break`, АБО змінну-прапорець, наприклад `running = True`.

````python
# Ваш код для завдання 22:

phone_book = {
    "Олена": "0671112233",
    "Андрій": "0502223344",
    "Марія": "0933334455",
    "Богдан": "0664445566",
    "Софія": "0955556677"
}

running = True

while running:
    print("\nДії: пошук / додати / видалити / вихід")
    action = input("Введіть дію: ").lower()

    if action == "пошук":
        name = input("Введіть ім'я: ")
        if name in phone_book:
            print(name, ":", phone_book[name])
        else:
            print("Контакт не знайдено")

    elif action == "додати":
        name = input("Ім'я: ")
        number = input("Телефон: ")
        phone_book[name] = number
        print("Контакт додано")

    elif action == "видалити":
        name = input("Ім'я: ")
        if name in phone_book:
            del phone_book[name]
            print("Контакт видалено")
        else:
            print("Контакт не знайдено")

    elif action == "вихід":
        running = False
        print("Програма завершена")

    else:
        print("Невідома дія")
````

**Приклад:** для `["кіт", "пес", "слон", "лев"]` → `{3: ["кіт", "пес", "лев"], 4: ["слон"]}`

> 💡 **Підказка:** Перевіряйте, чи є ключ-довжина у словнику. Якщо є — додайте слово до існуючого списку через `.append()`. Якщо немає — створіть новий ключ зі списком, у якому одне слово.

````python
words = ["кіт", "програма", "сонце", "пес", "комп'ютер", "стіл", "лев", "ніч", "школа", "день"]
# Ваш код:

result = {}

for word in words:
    length = len(word)

    if length in result:
        result[length].append(word)
    else:
        result[length] = [word]

print(result)
````

````python
# Ваш код для завдання 26:
history = []

while True:
    op = input("Введіть операцію (+, -, *, /) або 'вихід': ")

    if op == "вихід":
        break

    a = float(input("Перше число: "))
    b = float(input("Друге число: "))

    result = None

    if op == "+":
        result = a + b
    elif op == "-":
        result = a - b
    elif op == "*":
        result = a * b
    elif op == "/":
        if b != 0:
            result = a / b
        else:
            result = "Помилка (ділення на 0)"
    else:
        print("Невідома операція")
        continue

    record = f"{a} {op} {b} = {result}"
    history.append(record)

print("\nІсторія обчислень:")
for item in history:
    print(item)
````

````python
matches = [
    {"team_a": "Динамо",     "team_b": "Шахтар",     "score_a": 1, "score_b": 2},
    {"team_a": "Зоря",       "team_b": "Динамо",     "score_a": 0, "score_b": 1},
    {"team_a": "Шахтар",     "team_b": "Дніпро",     "score_a": 3, "score_b": 1},
    {"team_a": "Динамо",     "team_b": "Ворскла",    "score_a": 2, "score_b": 2},
    {"team_a": "Шахтар",     "team_b": "Зоря",       "score_a": 4, "score_b": 0},
    {"team_a": "Дніпро",     "team_b": "Ворскла",    "score_a": 1, "score_b": 1},
    {"team_a": "Зоря",       "team_b": "Дніпро",     "score_a": 2, "score_b": 1},
    {"team_a": "Ворскла",    "team_b": "Шахтар",     "score_a": 0, "score_b": 3}
]
# Ваш код:

points = {}

for match in matches:
    a = match["team_a"]
    b = match["team_b"]
    sa = match["score_a"]
    sb = match["score_b"]

    if a not in points:
        points[a] = 0
    if b not in points:
        points[b] = 0

    if sa > sb:
        points[a] += 3
    elif sa < sb:
        points[b] += 3
    else:
        points[a] += 1
        points[b] += 1

sorted_table = sorted(points.items(), key=lambda x: x[1], reverse=True)

print("Турнірна таблиця:")
for i, (team, pts) in enumerate(sorted_table, 1):
    print(f"{i}. {team} — {pts} очок")
````


````python
# Ваш код для завдання 41:

import random

user_wins = 0
computer_wins = 0

while user_wins < 3 and computer_wins < 3:
    user = input("Введіть: камінь, ножиці або папір: ").lower()

    if user not in ["камінь", "ножиці", "папір"]:
        print("Неправильний вибір")
        continue

    computer = random.choice(["камінь", "ножиці", "папір"])

    print("Комп'ютер вибрав:", computer)

    if user == computer:
        print("Нічия")

    elif (
        (user == "камінь" and computer == "ножиці") or
        (user == "ножиці" and computer == "папір") or
        (user == "папір" and computer == "камінь")
    ):
        print("Ви перемогли в раунді!")
        user_wins += 1

    else:
        print("Комп'ютер переміг у раунді!")
        computer_wins += 1

    print("Рахунок:", user_wins, "-", computer_wins)
    print()

if user_wins == 3:
    print("Вітаю! Ви виграли гру!")
else:
    print("Комп'ютер виграв гру!")
````

````python
# Ваш код для завдання 44:

alphabet = "абвгґдеєжзиіїйклмнопрстуфхцчшщьюя"

cipher = {}

for i in range(len(alphabet)):
    cipher[alphabet[i]] = alphabet[(i + 1) % len(alphabet)]

word = input("Введіть слово: ").lower()

result = ""

for ch in word:
    if ch in cipher:
        result += cipher[ch]
    else:
        result += ch

print("Зашифроване слово:", result)
````

Створіть змінну `balance = 5000`. У циклі `while` запитуйте у користувача дію:
- `баланс` — показати баланс
- `зняти` — запитати суму і зняти (якщо вистачає коштів)
- `поповнити` — запитати суму і додати до балансу
- `історія` — показати список усіх операцій
- `вихід` — завершити

Зберігайте кожну операцію у списку як словник `{"type": "зняти", "amount": 500}`.

> 💡 **Підказка:** Створіть `history = []` перед циклом. Кожну операцію додавайте через `.append()`.

````python
# Ваш код для завдання 45:

balance = 5000
history = []

while True:
    action = input("\nОберіть дію (баланс / зняти / поповнити / історія / вихід): ").lower()

    if action == "баланс":
        print("Поточний баланс:", balance)

        history.append({"type": "баланс", "amount": 0})

    elif action == "зняти":
        amount = float(input("Сума для зняття: "))

        if amount <= balance:
            balance -= amount
            print("Знято:", amount)
            history.append({"type": "зняти", "amount": amount})
        else:
            print("Недостатньо коштів")

    elif action == "поповнити":
        amount = float(input("Сума поповнення: "))
        balance += amount
        print("Поповнено:", amount)
        history.append({"type": "поповнити", "amount": amount})

    elif action == "історія":
        print("\nІсторія операцій:")
        for op in history:
            print(op)

    elif action == "вихід":
        print("Завершення роботи банкомату")
        break

    else:
        print("Невідома дія")
````

### Завдання 46. Менеджер задач (todo) ⭐⭐⭐

Створіть програму-менеджер задач. Задачі зберігаються у списку словників: `{"task": "...", "done": False}`. У циклі `while` запитуйте дію:
- `додати` — додати нову задачу
- `показати` — показати всі задачі з номерами і статусом
- `виконати` — за номером позначити задачу як виконану
- `видалити` — за номером видалити задачу
- `вихід` — завершити

````python
# Ваш код для завдання 46:

tasks = []

while True:
    action = input("\nОберіть дію (додати / показати / виконати / видалити / вихід): ").lower()

    if action == "додати":
        task_text = input("Введіть задачу: ")
        tasks.append({"task": task_text, "done": False})
        print("Задачу додано")

    elif action == "показати":
        if not tasks:
            print("Список задач порожній")
        else:
            print("\nСписок задач:")
            for i, t in enumerate(tasks, start=1):
                status = "✓" if t["done"] else "✗"
                print(f"{i}. {t['task']} [{status}]")

    elif action == "виконати":
        num = int(input("Введіть номер задачі: ")) - 1

        if 0 <= num < len(tasks):
            tasks[num]["done"] = True
            print("Задачу позначено як виконану")
        else:
            print("Невірний номер")

    elif action == "видалити":
        num = int(input("Введіть номер задачі: ")) - 1

        if 0 <= num < len(tasks):
            removed = tasks.pop(num)
            print("Видалено:", removed["task"])
        else:
            print("Невірний номер")

    elif action == "вихід":
        print("Завершення програми")
        break

    else:
        print("Невідома дія")
````

````python
network = {
    "Олена": {
        "friends": ["Андрій", "Марія", "Софія", "Петро"],
        "posts": [
            {"text": "Доброго ранку!", "likes": 25},
            {"text": "Нова стаття",    "likes": 48}
        ]
    },
    "Андрій": {
        "friends": ["Олена", "Богдан", "Софія"],
        "posts": [
            {"text": "Тренування",     "likes": 12},
            {"text": "Подорож",        "likes": 67}
        ]
    },
    "Марія": {
        "friends": ["Олена", "Дарʼя", "Петро", "Михайло", "Софія"],
        "posts": [
            {"text": "Книга",          "likes": 35}
        ]
    },
    "Богдан": {
        "friends": ["Андрій", "Михайло"],
        "posts": [
            {"text": "Робота",         "likes": 8}
        ]
    },
    "Софія": {
        "friends": ["Олена", "Андрій", "Марія"],
        "posts": [
            {"text": "Фото",           "likes": 92},
            {"text": "Музика",         "likes": 15},
            {"text": "Думки",          "likes": 28}
        ]
    },
    "Петро": {
        "friends": ["Олена", "Марія"],
        "posts": []
    },
    "Дарʼя": {
        "friends": ["Марія"],
        "posts": [
            {"text": "Готую",          "likes": 41}
        ]
    },
    "Михайло": {
        "friends": ["Богдан", "Марія"],
        "posts": [
            {"text": "Привіт",         "likes": 5}
        ]
    }
}
# Ваш код:

most_friends_user = ""
max_friends = 0

for user, data in network.items():
    count = len(data["friends"])
    if count > max_friends:
        max_friends = count
        most_friends_user = user

best_post_user = ""
best_post_text = ""
max_likes = 0

for user, data in network.items():
    for post in data["posts"]:
        if post["likes"] > max_likes:
            max_likes = post["likes"]
            best_post_text = post["text"]
            best_post_user = user

most_active_user = ""
max_total_likes = 0

for user, data in network.items():
    total_likes = 0
    for post in data["posts"]:
        total_likes += post["likes"]

    if total_likes > max_total_likes:
        max_total_likes = total_likes
        most_active_user = user

olena_friends = set(network["Олена"]["friends"])
andriy_friends = set(network["Андрій"]["friends"])

common_friends = olena_friends.intersection(andriy_friends)

print("Найпопулярніший користувач:", most_friends_user, "-", max_friends, "друзів")
print("Найпопулярніший пост:", best_post_text, "(", best_post_user, ")", "-", max_likes, "лайків")
print("Найактивніший користувач:", most_active_user, "-", max_total_likes, "лайків")
print("Спільні друзі Олени та Андрія:", list(common_friends))
````

### Завдання 52. Гра в слова ⭐⭐⭐

Реалізуйте гру: дано стартове слово. Гравець має ввести наступне слово, яке починається на ОСТАННЮ літеру попереднього. Якщо слово вже використовувалось — програш. Якщо слово починається не на ту літеру — програш. Цикл триває, поки гравець не програє або не введе `здаюсь`.

> 💡 **Підказка:** Зберігайте всі використані слова у списку. Першу і останню літеру отримуйте через індекси: `word[0]` і `word[-1]`.

````python
# Ваш код для завдання 52:

start_word = "кіт"

used_words = [start_word]
current_word = start_word

print("Стартове слово:", start_word)

while True:
    new_word = input("Наступне слово (або 'здаюсь'): ").lower()

    if new_word == "здаюсь":
        print("Гра завершена. Ви здались.")
        break

    if new_word in used_words:
        print("Програш! Це слово вже використовувалось.")
        break

    if new_word[0] != current_word[-1]:
        print("Програш! Невірна перша літера.")
        break

    used_words.append(new_word)
    current_word = new_word
    print("OK! Продовжуємо...")

print("\nВикористані слова:", used_words)
````

### Завдання 53. Аналіз журналу ⭐⭐⭐

Дано список словників-записів журналу логів. Кожен запис: `{"level": "INFO"|"WARNING"|"ERROR", "message": "...", "hour": ...}`. Створіть звіт:
1. Скільки записів кожного рівня (словник)
2. У яку годину було найбільше ERROR-ів
3. Усі повідомлення з рівнем ERROR
4. Чи були години без жодного ERROR (з тих, у яких є записи)

````python
logs = [
    {"level": "INFO",    "message": "Сервер запущено",     "hour": 8},
    {"level": "INFO",    "message": "Користувач увійшов", "hour": 9},
    {"level": "WARNING", "message": "Повільний запит",     "hour": 9},
    {"level": "ERROR",   "message": "Помилка БД",          "hour": 10},
    {"level": "ERROR",   "message": "Тайм-аут запиту",     "hour": 10},
    {"level": "INFO",    "message": "Запит виконано",      "hour": 11},
    {"level": "WARNING", "message": "Низько місця на диску","hour": 12},
    {"level": "ERROR",   "message": "Помилка БД",          "hour": 14},
    {"level": "INFO",    "message": "Резервна копія",      "hour": 15},
    {"level": "ERROR",   "message": "Збій авторизації",    "hour": 14},
    {"level": "ERROR",   "message": "Помилка БД",          "hour": 14},
    {"level": "INFO",    "message": "Користувач вийшов",   "hour": 17}
]
# Ваш код:

level_count = {}

error_by_hour = {}

error_messages = []

hours_with_logs = set()
hours_with_error = set()

for log in logs:
    level = log["level"]
    hour = log["hour"]

    if level in level_count:
        level_count[level] += 1
    else:
        level_count[level] = 1

    hours_with_logs.add(hour)

    if level == "ERROR":
        error_messages.append(log["message"])
        hours_with_error.add(hour)

        if hour in error_by_hour:
            error_by_hour[hour] += 1
        else:
            error_by_hour[hour] = 1

max_error_hour = None
max_errors = 0

for hour, count in error_by_hour.items():
    if count > max_errors:
        max_errors = count
        max_error_hour = hour

hours_without_error = hours_with_logs - hours_with_error

print("Кількість записів за рівнями:", level_count)
print("ERROR по годинах:", error_by_hour)
print("Година з найбільшою кількістю ERROR:", max_error_hour, "-", max_errors)
print("ERROR повідомлення:", error_messages)
print("Години без ERROR:", sorted(list(hours_without_error)))
````

### Завдання 54. Магазин з фільтрами ⭐⭐⭐

Дано список товарів — словники з полями `name`, `category`, `price`, `rating`, `in_stock`. У циклі `while` запитуйте у користувача фільтр:
- `категорія` — запитайте назву категорії, виведіть товари цієї категорії
- `ціна` — запитайте максимальну ціну, виведіть товари дешевше
- `рейтинг` — виведіть товари з рейтингом ≥ 4.5
- `в наявності` — виведіть товари, у яких `in_stock = True`
- `топ` — виведіть 3 найдорожчі товари
- `вихід` — завершити

````python
products = [
    {"name": "Ноутбук HP",     "category": "Електроніка", "price": 22000, "rating": 4.6, "in_stock": True},
    {"name": "Ноутбук Lenovo", "category": "Електроніка", "price": 18500, "rating": 4.3, "in_stock": True},
    {"name": "Телефон Samsung","category": "Електроніка", "price": 12000, "rating": 4.7, "in_stock": False},
    {"name": "Книга Python",   "category": "Книги",       "price": 450,   "rating": 4.8, "in_stock": True},
    {"name": "Книга SQL",      "category": "Книги",       "price": 380,   "rating": 4.2, "in_stock": True},
    {"name": "Кросівки Nike",  "category": "Одяг",        "price": 2800,  "rating": 4.5, "in_stock": True},
    {"name": "Куртка Adidas",  "category": "Одяг",        "price": 3500,  "rating": 4.0, "in_stock": False},
    {"name": "Навушники",      "category": "Електроніка", "price": 1200,  "rating": 4.6, "in_stock": True},
    {"name": "Чайник",         "category": "Побутова",    "price": 850,   "rating": 4.4, "in_stock": True},
    {"name": "Блендер",        "category": "Побутова",    "price": 1500,  "rating": 4.7, "in_stock": False}
]
# Ваш код:

while True:
    action = input("\nФільтр (категорія / ціна / рейтинг / в наявності / топ / вихід): ").lower()

    if action == "категорія":
        cat = input("Введіть категорію: ")
        for p in products:
            if p["category"].lower() == cat.lower():
                print(p["name"], "-", p["price"], "грн")

    elif action == "ціна":
        max_price = float(input("Максимальна ціна: "))
        for p in products:
            if p["price"] <= max_price:
                print(p["name"], "-", p["price"], "грн")

    elif action == "рейтинг":
        for p in products:
            if p["rating"] >= 4.5:
                print(p["name"], "-", p["rating"])

    elif action == "в наявності":
        for p in products:
            if p["in_stock"]:
                print(p["name"], "-", p["price"], "грн")

    elif action == "топ":
        top_products = sorted(products, key=lambda x: x["price"], reverse=True)[:3]
        for p in top_products:
            print(p["name"], "-", p["price"], "грн")

    elif action == "вихід":
        print("Завершення")
        break

    else:
        print("Невідома команда")
````
